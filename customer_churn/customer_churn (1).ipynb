```python
q1 = data['4 What is the Pin Code of where you shop online from?'].quantile(.25)
q3 = data['4 What is the Pin Code of where you shop online from?'].quantile(.75)
print("25 quartile of is:",q1)
print("75 quartile of is:",q3)

iqr = q3 - q1
print("The IQR value is:",iqr)

high_range = q3 + (1.5*iqr)  
low_range = q1 - (1.5*iqr)
print("The high value is",high_range)
print("The low value is",low_range)


data['4 What is the Pin Code of where you shop online from?'] = np.where(
    data['4 What is the Pin Code of where you shop online from?'] > high_range,
    high_range,
    np.where(
        data['4 What is the Pin Code of where you shop online from?'] < low_range,
        low_range,
        data['4 What is the Pin Code of where you shop online from?']
    )
)

data['4 What is the Pin Code of where you shop online from?'].plot(kind='box', subplots=True, layout=(4,7), figsize = (15,13))
```
